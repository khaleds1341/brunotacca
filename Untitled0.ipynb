```python
from datasets import load_dataset

# Load the dataset from Hugging Face
dataset = load_dataset("KFUPM-JRCAI/arabic-generated-abstracts")
split = dataset["by_polishing"]

# Collect real and fake abstracts
texts = []
labels = []

for row in split:
    # Real
    if row['original_abstract']:
        texts.append(row['original_abstract'])
        labels.append(0)

    # Generated
    for model_key in ['allam_generated_abstract', 'jais_generated_abstract',
                      'llama_generated_abstract', 'openai_generated_abstract']:
        if row[model_key]:  # Check for missing values
            texts.append(row[model_key])
            labels.append(1)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split

df = pd.DataFrame({'text': texts, 'label': labels})

# Split into train/test
train_texts, test_texts, train_labels, test_labels = train_test_split(
    df['text'], df['label'], test_size=0.2, random_state=42, stratify=df['label']
)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report

# Arabic tokenizer can be added later โ this is basic.
model = Pipeline([
    ("tfidf", TfidfVectorizer(max_features=5000)),
    ("clf", LogisticRegression(max_iter=1000))
])

# Train
model.fit(train_texts, train_labels)

# Predict
preds = model.predict(test_texts)

# Evaluation
print(classification_report(test_labels, preds))
```
